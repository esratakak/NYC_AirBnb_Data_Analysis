```python
df_nyc["neighbourhood_group"].value_counts().plot.barh().set_title("FOR NEİGHBOURHOOD")
```

```python
sns.barplot(x="neighbourhood_group", y="price", data=df_nyc).set_title("FOR NEİGHBOURHOOD")
```

```python
sns.barplot(x="room_type", y="price", data=df_nyc).set_title("FOR ROOM TYPE")
```

```python
sns.barplot(x = 'neighbourhood_group', y = 'price', hue = 'room_type', data = df_nyc)
plt.title("Neighbourhood - Price - Color")
```

```python
df_nyc.groupby("neighbourhood_group").sum().plot.pie(y="price", autopct="%.1f%%", ylabel="", legend=False, figsize=(5,5))
```

```python
df_nyc.groupby("room_type").sum().plot.pie(y="price", autopct="%.1f%%", ylabel="", legend=False, figsize=(5,5))
```

```python
cols = ["room_type", "neighbourhood_group", "number_of_reviews"]

for col in cols:
  plt.figure(figsize=(5,5))
  sns.scatterplot(x=col, y="price", data=df_nyc)
```

```python
sns.histplot(df_nyc["price"], kde=True)
```

```python
sns.kdeplot(df_nyc["price"], shade= True)
```

```python
sns.jointplot(x="price", y=df_nyc["neighbourhood_group"], data=df_nyc)
```

```python
sns.catplot(x="price", y="neighbourhood_group", data=df_nyc)
plt.title("Neighbourhoods - Price")
```

```python
sns.catplot(x="price", y="room_type", data=df_nyc)
plt.title("Room - Price")
```

```python
sns.catplot(x="room_type", y="price", hue="neighbourhood_group", kind="point", data=df_nyc)
plt.title("Cut - Price")
```

```python
plt.title("Correlation Matrix")
sns.heatmap(df_nyc.corr(), annot=True);
```
