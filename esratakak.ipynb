```python
sns.barplot(x="neighbourhood_group", y=df["neighbourhood_group"].index, data=df, color= "grey").set_title("FOR NEİGHBOURHOOD")
```

```python
sns.barplot(x = 'neighbourhood_group', y = 'price', hue = 'room_type', data = df)
plt.title("Neighbourhood - Price - Color")
```

```python
sns.barplot(x="neighbourhood_group",
            y="price",
            hue="number_of_reviews",
            data=df)
plt.title("Neighbourhood - Price - Available")
```

```python
df.groupby("neighbourhood_group").sum().plot.pie(y="price", autopct="%.1f%%", ylabel="", legend=False, figsize=(5,5))
```
